```python
from matplotlib import pyplot as plt
import random
```

```python
x = range(120)
y = [random.randint(20,35) for _ in range(120)]
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(20,8), dpi=80)

plt.plot(x, y)
 
x_ticks = ["10点{}分".format(i) for i in range(0, 60)]
x_ticks += ["11点{}分".format(i) for i in range(60,120)]

plt.xticks(x[::3], x_ticks[::3],rotation=45)

plt.xlabel("时间")
plt.ylabel("温度")
plt.title("10点到12点的温度变化")

plt.show()
```
