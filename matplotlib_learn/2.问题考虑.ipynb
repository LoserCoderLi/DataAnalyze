```python
from matplotlib import pyplot as plt
import random
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
x = (range(2, 26, 2))
y = [random.randint(13, 27) for i in x]
```

```python
y_max = np.argmax(y)
y_min = np.argmin(y)
```

```python
plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y,color='orange',label="22号")
_x = [i*0.5 for i in range(4,49)]
x_ticks = ["{}时".format(i) for i in _x]
plt.xticks(_x, x_ticks, rotation=45)
plt.xlabel("时间")
plt.ylabel("温度")

plt.legend(loc="upper left")

plt.scatter(x[y_max], y[y_max], color='blue')
plt.scatter(x[y_min], y[y_min], color='red')

plt.text(x[y_max], y[y_max],"最大值")
plt.savefig("./output.png")



plt.show()
```
