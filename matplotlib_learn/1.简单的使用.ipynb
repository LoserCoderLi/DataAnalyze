```python
from matplotlib import pyplot as plt
import random
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
x = (range(2, 26, 2))
y = [random.randint(13, 27) for i in x]
```

```python
# x
# y
```

```python
plt.plot(x, y)

_x = x
x_lable = ["{}小时".format(i) for i in _x]

plt.xticks(x, x_lable, rotation=90)

plt.show()
```
