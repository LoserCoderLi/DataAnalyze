```python
from matplotlib import pyplot as plt
import random
```

```python
a = ["猩球崛起3:终极之战", "敦刻尔克", "蜘蛛侠:英雄归来", "战狼2"]

b16=[15746,312,4497,319]
b15=[12357,156,2045,168]
b14 =[2358,399,2358,362]
```

```python
x16 = list(range(len(a)))
x15 = [i + 0.2 for i in x16]
x14 = [i + 0.2*2 for i in x16]

plt.rcParams['font.sans-serif'] = ["SimHei"]

plt.figure(figsize=(20,8), dpi=80)
plt.barh(x16, b16, height=0.2, label="2017-9-16")
plt.barh(x15, b15, height=0.2, label="2017-9-15")
plt.barh(x14, b14, height=0.2, label="2017-9-14")

plt.yticks(x15, a)

plt.legend()
plt.grid(True)

plt.xlabel("票房数量")
plt.ylabel("电影")

plt.show()
```
