```python
x_3 = list(range(1,32))
x_10 = list(range(1,32))

y_3=[11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
y_10=[26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6]

plt.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(20,8),dpi=80)
_x_10 = [i+31 for i in x_10]

plt.scatter(x_3, y_3, label="三月")
plt.scatter(_x_10, y_10, label="三月")

plt.legend(loc='upper left')

plt.xlabel("月份")
plt.ylabel("温度")

_x_3_ticks = ["3月{}号".format(i) for i in x_3]
_x_3_ticks += ["10月{}号".format(i) for i in x_10]

print(_x_3_ticks)

_x = x_3 + _x_10

plt.xticks(_x[::2], _x_3_ticks[::2], rotation=45)
# plt.xticks(_x_10, _x_10_ticks, rotation=45)

# plt.yticks(y_3[::1])

plt.show()
```
